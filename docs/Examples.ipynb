# Examples of using coolpup.py
To reduce the file size in the repository and allow quick computation, here I demonstrate some options of coolpup.py using pooled single-nucleus Hi-C data from Gassler et al., 2017 paper in EMBO Journal.

```bash
cd docs
```

First let's create pileups of loops in control cells and cells with Scc1 (aka Rad21) gene knocked-out, which should disrupt cohesin activity and abrogate loop extrusion

```bash
coolpup.py Scc1-control.10000.cool CH12_loops_Rao.bed --nshifts 10 --unbalanced --coverage_norm --mindist 100000 --outname Ctrl_loop.txt --log WARNING #Suppressing long default outputs
```

```bash
coolpup.py Scc1-KO.10000.cool CH12_loops_Rao.bed --nshifts 10 --unbalanced --coverage_norm --mindist 100000 --outname Scc1KO_loop.txt --log WARNING
```

Now that we have the results, let's plot them using plotpup.py

```bash
plotpup.py Ctrl_loop.txt Scc1KO_loop.txt --row_names Control,Scc1KO --enrichment 3 --output loops.png
```

```bash
display <loops.png
```

Now let\'s repeat the same procedure, but with local rescaled pileups of TADs

```bash
coolpup.py Scc1-control.10000.cool CH12_TADs_Rao.bed --nshifts 10 --rescale --local --unbalanced --coverage_norm --rescale_size 87 --outname Ctrl_TAD.txt --log WARNING
```

```bash
coolpup.py Scc1-KO.10000.cool CH12_TADs_Rao.bed --nshifts 10 --unbalanced --coverage_norm --local --rescale --rescale_size 87 --outname Scc1KO_TAD.txt --log WARNING
```

```bash
plotpup.py Ctrl_TAD.txt Scc1KO_TAD.txt --row_names Control,Scc1KO --enrichment 0 --output TADs.png
```

```bash
display <TADs.png
```

Now let\'s plot all together to demonstrate how plotting in noth columns and rows works. plotpup.py will determine the number of columns and rows automatically to make the figure as square as possible, but you could instead specify --n_cols to set the number of columns (which determines the number of rows)
Note that plotting is done along columns: we first specify two control pileups, and they are shown in the top row.

```bash
plotpup.py Ctrl_loop.txt Ctrl_TAD.txt Scc1KO_loop.txt Scc1KO_TAD.txt --col_names Loop,TAD --row_names Control,Scc1KO --enrichment 0 --output all.png
```

```bash
display < all.png
```
